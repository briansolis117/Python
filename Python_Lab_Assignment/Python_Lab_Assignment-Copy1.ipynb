```python
# Dependencies and Setup
import pandas as pd
import numpy as np
import sqlite3

# Loading data from python.db database
con = sqlite3.connect('data/python.db')

# Import 'school' and 'student' tables into pandas DataFrames
school_data = pd.read_sql_query("SELECT * FROM school", con)
student_data = pd.read_sql_query("SELECT * FROM student", con)

# Close the connection
con.close()

# Display first few rows of each DataFrame to verify
print(school_data.head())
print(student_data.head())

```

```python
# join the two tables into a single dataframe
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Create a District Summary
# Calculate the total number of schools
total_schools = school_data["school_name"].nunique()

# Calculate the total number of students
total_students = school_data_complete["student_name"].count()

# Calculate the total budget
total_budget = school_data["budget"].sum()

# Calculate the average math score
avg_math_score = school_data_complete["math_score"].mean()

# Calculate the average reading score
avg_reading_score = school_data_complete["reading_score"].mean()

# Calculate the overall passing rate
overall_passing_rate = (avg_math_score + avg_reading_score) / 2

# Calculate the percentage of students with a passing math score (70 or greater)
passing_math_percentage = (
    (school_data_complete["math_score"] >= 70).sum() / total_students
) * 100

# Calculate the percentage of students with a passing reading score (70 or greater)
passing_reading_percentage = (
    (school_data_complete["reading_score"] >= 70).sum() / total_students
) * 100

# Create a DataFrame to hold the results
district_summary = pd.DataFrame(
    {
        "Total Schools": [total_schools],
        "Total Students": [total_students],
        "Total Budget": [total_budget],
        "Average Math Score": [avg_math_score],
        "Average Reading Score": [avg_reading_score],
        "Overall Passing Rate": [overall_passing_rate],
        "% Passing Math": [passing_math_percentage],
        "% Passing Reading": [passing_reading_percentage],
    }
)

# Optional: format the output for better readability
district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)
district_summary["Average Math Score"] = district_summary["Average Math Score"].map("{:.2f}".format)
district_summary["Average Reading Score"] = district_summary["Average Reading Score"].map("{:.2f}".format)
district_summary["Overall Passing Rate"] = district_summary["Overall Passing Rate"].map("{:.2f}".format)
district_summary["% Passing Math"] = district_summary["% Passing Math"].map("{:.2f}%".format)
district_summary["% Passing Reading"] = district_summary["% Passing Reading"].map("{:.2f}%".format)

# Display the District Summary
district_summary


```

### Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
#  Sort and display the top five schools in overall passing rate
# Calculate key metrics for each school
school_summary = school_data_complete.groupby("school_name").agg(
    total_students=("student_name", "count"),
    total_budget=("budget", "first"),
    avg_math_score=("math_score", "mean"),
    avg_reading_score=("reading_score", "mean"),
    passing_math_percentage=("math_score", lambda x: (x >= 70).sum() / len(x) * 100),
    passing_reading_percentage=("reading_score", lambda x: (x >= 70).sum() / len(x) * 100)
)

# Calculate per student budget
school_summary["per_student_budget"] = school_summary["total_budget"] / school_summary["total_students"]

# Calculate overall passing rate
school_summary["overall_passing_rate"] = (
    (school_summary["passing_math_percentage"] + school_summary["passing_reading_percentage"]) / 2
)

# Optional: format the output for better readability
school_summary["total_budget"] = school_summary["total_budget"].map("${:,.2f}".format)
school_summary["per_student_budget"] = school_summary["per_student_budget"].map("${:,.2f}".format)
school_summary["avg_math_score"] = school_summary["avg_math_score"].map("{:.2f}".format)
school_summary["avg_reading_score"] = school_summary["avg_reading_score"].map("{:.2f}".format)
school_summary["passing_math_percentage"] = school_summary["passing_math_percentage"].map("{:.2f}%".format)
school_summary["passing_reading_percentage"] = school_summary["passing_reading_percentage"].map("{:.2f}%".format)
school_summary["overall_passing_rate"] = school_summary["overall_passing_rate"].map("{:.2f}%".format)

# Display the summary
school_summary.head()

# Sort by overall passing rate and display top 5 schools
top_performing_schools = school_summary.sort_values("overall_passing_rate", ascending=False).head(5)
top_performing_schools

```

#### Find the passing rate for math and reading (above 70 points)

```python
# Find the total counts of math result
# Total count of math results (total number of students)
total_math_count = school_data_complete["math_score"].count()
print("Total Math Results (Students):", total_math_count)


# Find the counts for math result in each school that pass 70 or higher
# Count of students passing math (math score >= 70) for each school
passing_math_count = school_data_complete[school_data_complete["math_score"] >= 70].groupby("school_name")["math_score"].count()

# Display the passing math counts for each school
passing_math_count


# Calculate the math passing rate
# Calculate the passing math rate for each school
math_passing_rate = (passing_math_count / school_data_complete.groupby("school_name")["math_score"].count()) * 100

# Display the math passing rate for each school
math_passing_rate


```

```python
# Find the total counts of read result
# Total count of reading results (total number of students)
total_read_count = school_data_complete["reading_score"].count()
print("Total Reading Results (Students):", total_read_count)


# Find the counts for read result in each school that pass 70 or higher
# Count of students passing reading (reading score >= 70) for each school
passing_read_count = school_data_complete[school_data_complete["reading_score"] >= 70].groupby("school_name")["reading_score"].count()

# Display the passing reading counts for each school
print(passing_read_count)


# Calculate the read passing rate
# Calculate the passing reading rate for each school
read_passing_rate = (passing_read_count / school_data_complete.groupby("school_name")["reading_score"].count()) * 100

# Display the reading passing rate for each school
print(read_passing_rate)

```

```python
# Calculate the overall passing rate (average of the math and reading passing rate)
# Calculate the math passing rate for each school
math_passing_rate = (passing_math_count / school_data_complete.groupby("school_name")["math_score"].count()) * 100

# Calculate the reading passing rate for each school
read_passing_rate = (passing_read_count / school_data_complete.groupby("school_name")["reading_score"].count()) * 100

# Calculate the overall passing rate (average of math and reading passing rates)
overall_passing_rate = (math_passing_rate + read_passing_rate) / 2

# Display the overall passing rate for each school
print(overall_passing_rate)


```
